```python
import matplotlib.pyplot as plt
%matplotlib notebook

def beregning(CON, prof, advantage = False, dcListe = [i for i in range(0, 31)]):
    resultatliste = []
    for dc in dcListe:
        minimumRoll = dc - CON - prof
        sannsynlighet = 100 - 100*(minimumRoll-1)/20
        if sannsynlighet > 100:
            sannsynlighet = 100
        elif sannsynlighet < 0:
            sannsynlighet = 0
        if advantage:
            sannsynlighet = 100 - 100*(1 - sannsynlighet/100)**2
        
        resultatliste.append(int(sannsynlighet))
    return(resultatliste)

def plotting():
    resultatmatrise = []
    labelListe = []
    dcListe = [i for i in range(0, 31)]
    for CON in range(3,4):
        for prof in range(4,5):
            resultatmatrise.append(beregning(CON, prof))
            resultatmatrise.append(beregning(CON, 0, True))
            labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=False')
            labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=True')
    for i in range(len(resultatmatrise)):
        plt.plot(dcListe, resultatmatrise[i], label=labelListe[i])
    plt.legend()
    plt.grid()
    plt.show()

def plotting2():
    resultatmatrise = []
    labelListe = []
    dcListe = [i for i in range(0, 31)]
    prof = 4
    CON = 4
    resultatmatrise.append(beregning(CON, prof))
    labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=False')
    CON = 3
    resultatmatrise.append(beregning(CON, 0, True))
    labelListe.append('CON='+str(CON)+', prof=0, Adv=True')
    for i in range(len(resultatmatrise)):
        plt.plot(dcListe, resultatmatrise[i], label=labelListe[i])
    plt.legend()
    plt.grid()
    plt.show()

def plotting3():
    resultatmatrise = []
    labelListe = []
    dcListe = [i for i in range(0, 31)]
    prof = 4
    CON = 4
    resultatmatrise.append(beregning(CON, prof))
    labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=False')
    CON = 3
    resultatmatrise.append(beregning(CON, 0, True))
    labelListe.append('CON='+str(CON)+', prof=0, Adv=True')
    CON = 4
    resultatmatrise.append(beregning(CON, prof, True))
    labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=True')
    CON = 5
    resultatmatrise.append(beregning(CON, prof, False))
    labelListe.append('CON='+str(CON)+', prof='+str(prof)+', Adv=False')
    for i in range(len(resultatmatrise)):
        plt.plot(dcListe, resultatmatrise[i], label=labelListe[i])
    plt.legend()
    plt.grid()
    plt.show()

plotting2()
```
